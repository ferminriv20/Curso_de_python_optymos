## Ciclo for


---



Los ciclos "for" en Python son herramientas fundamentales para la iteración sobre secuencias de datos. Estos proporcionan una forma eficiente de realizar tareas repetitivas, permitiendo automatizar procesos y facilitar la manipulación de datos. En esta lección, exploraremos la sintaxis y el funcionamiento de los ciclos "for" en Python.

La sintaxis básica de este ciclo es la siguiente

````python
# Sintaxis básica de un for()
for variable in secuencia:
  # Código a ejecutar en cada iteración
````







**Función range**


---


 La función range es una estructura de datos que nos permite generar secuencias inmutables de numeros. Es extremadamente útil cuando trabajamos con secuencias numericas extensas, debido a que permite generar secuencias númericas eficentes sin almacenar los valores en la memoria.
La sintaxis básica para crear un rango es mediante la función **range(start, end, step)**, que toma uno, dos o tres argumentos. Los argumentos especifican el inicio, el fin (exclusivo) y el paso de la secuencia.
Se reitera la importancia de entender los parametros de la función range para su funcionamiento.


*   **start** : Corresponde al primer valor del rango, en caso de que solo se especifique un solo valor para range se asume que start = 0.
*   **end** : Hace referencia al limite superior del rango.Es importante tener en cuenta que este valor no se encuentra incluido.
*   **step** : Controla el incremento de los valores en el rango, por defecto step = 1.


````python
#Ejemplo #1
for i in range(20):
  print(i, end = ",")
````

````python
# Ejemplo #2
# imprima los cuadrados de los primeros 10 numeros
for i in range(11):
  print(i*2)
````

````python
#Imprime los primeros 100 números pares
for i in range(2,101,2):
  print(i)
````

````python
#Imprime los primeros 100 números impares
for i in range(1,100,2):
  print(i)
````

````python
# Muestre en pantalla los multiplos del 5 hasta el 1000 en orden INVERSO.
for i in range(1000,4,-5):
  print(i)
````



Es común usar el bucle for para iterar los elementos de una lista, la sintaxis generalizada para este caso es la siguiente :




````python
for VARIABLE  in LIST:
  BODY
````

````python
#Ejemplos
friends = ["Joe", "Zoe", "Brad", "Angelina", "Zuki", "Thandi", "Paris"]
#Forma #1
for amigo in friends:
  print("Hola " + amigo+ " quiero invitarte a mi fiesta de cumpleaños")
````

````python
#Forma #2
for amigo in friends:
  print(f"Hola {amigo} te invito a mi fiesta de cumpleaños")
````

````python
#Forma #3
for i in range(len(friends)):
  print(f"Hola {friends[i]} te invito a mi fiesta de cumpleaños")
````

A veces es útil  considerar el indice y el valor del elemento de la lista, cada par de estos valores recorriendo toda la lista puede ser generado  por medio de la función enumerate.

````python
#Modificar valores de una lista
xs = [1, 2, 3, 4, 5]
for (i, val) in enumerate(xs):
    xs[i] = val**2
print(xs)
````

````python
#Mostrar indices y elementos de una lista
for indice, elemento,  in enumerate(friends):
  print( indice, elemento )
````

````python
#Iteración sobre un string
name = input("Ingrese su nombre : ")
for i in name:
  print(i)
````

````python
#iteración sobre un diccionario
mi_dic = {
    'Fermin':1,
    'Antonio':2,
    'Rivero':3,
    'Sotelo':4
}
#Recorriendo las keys
for i in mi_dic.keys():
  print(i)
#Recorriendo values
for i in mi_dic.values():
  print(i)

#Recorriendo keys y values
for clave,valor in mi_dic.items():
  print(clave,"->", valor)
````

````python
# Tambien podemos usar más de un bucle for.
#Imprimiendo tablas de multiplicar
for i in range(1,10):
  for j in range(1,10):
    print(f"{i}x{j} = {i*j}")
````

````python
#ciclos con condicionales
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
  if numero % 2 == 0:
    print('Número par: ', numero)
  else:
    print('Número impar: ', numero)
````

## Ciclo While

---



Los ciclos while  en Python proporcionan una estructura de control que permite repetir un bloque de código mientras una condición sea verdadera. A diferencia de los ciclos for, suele utilizarse cuando no se conoce de antemano el número de repeticiones del programa  y pueden ejecutarse indefinidamente hasta que la condición se vuelva falsa.

La sintaxis básica del cliclo while es la siguiente

````python
while condicion:
  #codigo a ejecutar
````

````python
#Ejemplo simple
i = 0
while i < 10 :
  print(i)
  i= i+1
````

````python
# Ejemplo simple de otra forma
i = 10
while i >0 :
  print(i)
  i = i-1
````

````python
#Analicemos un poco que hace este programa...
i = 0
while i < 10 :
	if i%2==0 :
		print(i**2)
		i = i+1
	else :
		print(i**3)
		i = i+1
````

# Break, continue


---

Las declaraciones break, continue  son herramientas fundamentales en Python para controlar el flujo de ejecución en bucles y condicionales.


*  break: La declaración break se utiliza en Python para salir inmediatamente de un bucle o condicional interrumpiendo su ejecución. Cuando se encuentra la instrucción break, el control del programa se transfiere fuera del bucle, y la ejecución continúa con la primera instrucción después del bucle.
*   continue: La declaración continue se utiliza para pasar a la siguiente iteración de un bucle, omitiendo el resto del código dentro del bucle para la iteración actual. Cuando se encuentra la instrucción continue, el programa salta directamente al comienzo del bucle para la siguiente iteración.

Nota : adicionalmente existe la declaración pass, se sugiere leer la documentacion acerca de esta.









````python
# Ejemplo sentencia break
name = ["ana" , "Juan", "Abel", "Fhara", "Andres"]
for i in name:
  if i == "Fhara":
    print("Valor encontrado!! ")
    break
  print(i)
````

````python
# Ejemplo sentencia continue
name = ["ana" , "Juan", "Abel", "Fhara", "Andres"]
for i in name:
  if i == "Fhara":
    continue
  print(i)
````

````python
#Veamos un ejemplo más interesante..
import random
num_oculto = random.randint(1,6) #simulamos el lanzamiento de un dado
num_user = int(input("Adivina el numero : "))
while num_oculto != num_user :
  num_user = int(input("Adivina el numero : "))
print("Felicidades, adivinaste el número!!")
````

````python
# Juego de dados con un ciclo while y condicional
import random

puntuacion_objetivo = 20
puntuacion_jugador = 0

while puntuacion_jugador < puntuacion_objetivo:
    lanzamiento = random.randint(1, 6)
    print(f"Has lanzado un {lanzamiento}")

    if lanzamiento == 1:
        print("¡Perdiste! Sacaste un 1.")
        break  # El juego termina si se saca un 1
    else:
        puntuacion_jugador += lanzamiento
        print(f"Puntuación actual: {puntuacion_jugador}")

print("¡Ganaste!")
````

Las comprensiones o comprehensions  permiten a los desarrolladores escribir código más legible y conciso al mismo tiempo que fomentan una mentalidad orientada a expresar la lógica de programación de manera directa y elegante. En lugar de bucles tradicionales que requieren varias líneas de código, las comprensiones condensan esa lógica en una única expresión, lo que hace que el código sea más fácil de entender y mantener. Generalmente, utilizamos la comprensión en listas, diccionarios y conjuntos.



*   List comprehensions.

```
# lista = [expression for variable in iterable ]
```




````python
# Ejemplo
num = [1,3,5,7,9,11]
num_pares = [2*i for i in num]
print(num_pares)
````

````python
#Tradicionalmente el ejemplo anterior se realiza de esta forma.
num_pares = []
for i in num:
  num_pares.append(2*i)
print(num_pares)
````

* Dictionary comprehensions





```
# diccionario =  {clave: valor for elemento in iterable }.
```



````python
# Ejemplo #1
nombres = ['Alice', 'Bob', 'Charlie']
edades = [27 , 34, 24]
id = {nombre: edad for nombre, edad in zip(nombres, edades)}
print(id)
````

````python
#Ejemplo #2
nombres = ['Fermín', 'Lucas', 'NelsonM']
names = {nombre : len(nombre) for nombre in nombres}
print(names)
````

* Set comprehensions



```
# new_set = {expression for variable in iterable }
```



````python
#Ejemplo
data = [2,3,6,8,94,-1,5,7,3,6,3,8]
set_cubos = {i**3 for i in data}
print(set_cubos)
````

* Tuple comprehensions

````python
# Comprensión de lista que genera una tupla
mi_tupla = tuple(x**2 for x in range(5))
print(mi_tupla)
````

Adicionalmente, existe la posibilidad de crear comprehensions con condicionales para cada una de las estructuras de datos mencionadas, acontinuación se muestran algunos ejemplos.



```
# Sintaxis básica de comprensión con condicional
# lista = [ expresion for elemento in iterable if condicion ]
```





```
#  Sintaxis básica de comprension condicionada de diccionarios

diccionario = { llave:valor for llave, valor in pares_llave_valor if condicion}

diccionario = {}
for llave, valor in pares_llave_valor:
  if condicion:
    diccionario[ llave ] = valor
```



````python
# Ejemplo: Crear una lista de cuadrados solo para números pares
num= [1, 2, 3, 4, 5, 6]
numeros_pares =[i for i in num if i%2 ==0]
print(numeros_pares)
````

````python
# Ejemplo: Crear un diccionario de cuadrados solo para números mayores que 2
numeros = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
cuadrados_mayores_que_dos = {llave: valor**2 for llave, valor in numeros.items() if valor > 2 and llave !='c'}

cuadrados_mayores_que_dos_classic = {}
for llave, valor in numeros.items():
  if valor > 2 and llave !='c':
    cuadrados_mayores_que_dos_classic[ llave ] = valor**2

print(cuadrados_mayores_que_dos)
print(cuadrados_mayores_que_dos_classic)
````

Las comprensiones de listas son una característica notable de Python que permite construir listas de manera concisa y elegante. Utilizan una sintaxis compacta que a menudo reduce varias líneas de código a una sola expresión. Además de su legibilidad, las comprensiones de listas tienden a ser más eficientes en términos de rendimiento en comparación con los bucles tradicionales. Esto se debe a que las comprensiones de listas están optimizadas internamente en el intérprete de Python y se ejecutan en el nivel de C. Al evitar la necesidad de crear una lista vacía, iterar sobre elementos y agregarlos uno por uno, las comprensiones de listas pueden construir listas de manera más rápida y eficiente.
Finalmente, se muestra un ejemplo para evidenciar  el argumento anterior.

````python
import time

# ciclo For
inicio = time.time()

numbers = range(100000)
result = []
for num in numbers:
    if num % 3 - 1 == 1:
        result.append(num ** 5)
fin = time.time()
for_loop_time = fin - inicio

# List Comprehension
inicio = time.time()
result = [num ** 5 for num in numbers if num % 3 - 1 == 1]
fin = time.time()
list_comp_time = fin - inicio

print("Ciclo For:", for_loop_time)
print("List Comprehension:", list_comp_time)
print("Diferencia:", for_loop_time - list_comp_time)
````
